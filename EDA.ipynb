```python
# 3) Missing values
df=pd.read_csv('/content/creditcard 2.csv')
missing = df.isna().sum().sort_values(ascending=False)
print("\nMissing Values:\n", missing)
```

```python
# 4) Duplicate check
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# 5) Separate numeric & categorical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print("\nNumeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# Optional: Extract brand from CarName
if "CarName" in df.columns:
    df["brand"] = (
        df["CarName"]
        .astype(str)
        .str.strip()
        .str.split(" ")
        .str[0]
        .str.lower()
        .str.replace("-", "", regex=False)
        .str.replace(".", "", regex=False)
    )
    if "brand" not in categorical_cols:
        categorical_cols.append("brand")
```

```python
# 6) Descriptive statistics for numeric columns
print("\nNumeric Descriptive Statistics:\n", df[numeric_cols].describe().T)
```

```python
# 1. Correlation Matrix + Heatmap
df=pd.read_csv('/content/creditcard 2.csv')
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()

# Optional: correlations with target 'Class'
if 'Class' in df.columns:
    target_corr = corr_matrix['Class'].sort_values(ascending=False)
    print("Correlation with 'Class':\n", target_corr)

# 2. Pairplots (only for a subset of numerical columns to avoid heavy plotting)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
sample_cols = num_cols[:5]  # change number to view more/less
sns.pairplot(df[sample_cols], diag_kind="kde")
plt.suptitle("Pairplot for Selected Numerical Features", y=1.02)
plt.show()

# 3. Scatter plots & trend lines (example: Amount vs. V1)
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Amount', y='V1', data=df, alpha=0.5)
sns.regplot(x='Amount', y='V1', data=df, scatter=False, color='red')  # trend line
plt.title("Scatter Plot with Trend Line (Amount vs V1)")
plt.show()

# 4. Group-by statistics (example: group by Class)
if 'Class' in df.columns:
    group_means = df.groupby('Class').mean(numeric_only=True)
    print("\nGroup-by Mean Values by Class:\n", group_means)

    # Bar plot for group means of Amount
    plt.figure(figsize=(6, 4))
    group_means['Amount'].plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title("Mean Transaction Amount by Class")
    plt.ylabel("Mean Amount")
    plt.show()
```

```python
#Target Variable Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=df, palette="Set2")
plt.title("Class Distribution (0 = Non-Fraud, 1 = Fraud)")
plt.show()

fraud_pct = (df["Class"].value_counts(normalize=True) * 100).round(2)
print("\nFraud Percentage:\n", fraud_pct)
```

```python
# Distribution of Transaction Amounts
plt.figure(figsize=(8,5))
sns.histplot(df["Amount"], bins=50, kde=True, color="blue")
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show()

# Compare Fraud vs Non-Fraud Amounts
plt.figure(figsize=(8,5))
sns.boxplot(x="Class", y="Amount", data=df, palette="Set1")
plt.ylim(0, 500)  # zoom in (to avoid extreme outliers hiding the pattern)
plt.title("Amount by Class")
plt.show()
```

```python
# Feature Distributions
df.iloc[:,1:5].hist(figsize=(10,8), bins=50, color="purple")
plt.suptitle("Distribution of First 4 Features (V1-V4)")
plt.show()
```

```python
# Fraud vs Non-Fraud in Time
plt.figure(figsize=(8,5))
sns.histplot(df[df["Class"]==0]["Time"], bins=50, color="green", label="Non-Fraud", alpha=0.6)
sns.histplot(df[df["Class"]==1]["Time"], bins=50, color="red", label="Fraud", alpha=0.6)
plt.legend()
plt.title("Transaction Time Distribution (Fraud vs Non-Fraud)")
plt.xlabel("Time")
plt.show()
```
